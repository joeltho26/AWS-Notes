### AWS CLI profiles:
- To manage multiple AWS account during command line access to AWS resources
- open command line and type
    - aws configure --profile "profile name"
        - profile name => my-other-aws-account
    - aws s3 ls --profile my-other-aws-account

Note:
- if we don't use --profile in the command then it uses the default CLI

### AWS CLI for MFA users:
- we need temporary session for MFA with CLI
    - run STS GetSessionToken API call

#### Command:

aws sts get-session-token --serial-number "arn-of-the-mfa-device" --token-code "code-from-token" --duration-seconds "3600"

- After executing the command, we get the credentials and we can create an AWS CLI profile on MFA
