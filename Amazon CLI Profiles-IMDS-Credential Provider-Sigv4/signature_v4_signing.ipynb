#### AWS Signature v4 Signing (Sigv4):
- When we call AWS HTTP API i.e. for any AWS service, we sign the request so that AWS service can identify us for validating permissions. 
    - It is done by using AWS credentials (access key & secret key)
- Some request to AWS S3 do not be signed (public url)
- If we use AWS SDK or CLI, the HTTP requests are signed automatically
- We should sign AWS HTTP request using Signature v4 (SigV4)

#### SigV4 Request:

The HTTP request is signed in two ways:
- HTTP Header (Authorization Header)
- Query String (signature in X-Amz-Signature)
    - e.g.: presigned S3 URL
