#### Amazon Elastic Beanstalk:
- Automate all infrastructure such as EC2, ELB, RDS, ASG, etc
- Developer centric view of deploying apps
- Managed service
    - automatically handles capacity provisioning, load balancing, scaling, application health monitoring, instance configuration, etc
- just application code as developer responsibility
- free service but pay for EC2 instances
- We still have full control over the configuration of AWS infrastructure

#### Elastic Beanstalk Deployment Modes:
Deployment options for Updates:
- All at once: 
    - fastest, but no instance to handle the traffic and downtime in application
- Rolling: 
    - few instances at a time (bucket) and move on to the other bucket
- Rolling with batch: 
    - like rolling, but spins up new instances to move the batch (old application is still available)
- immutable:
    - spins up new instances via ASG and deploys new versions to the new instances and replacing the old ones
- Blue green:
    - create a new environment and switch over when ready
- Traffic Splitting:
    - canary testing, sends a small % of traffic to new deployment

All at Once:
- Fastest deployment
- Downtime
- Quick iterations
- No additional costs

Traffic Splitting:
- Canary testing
    - a software testing technique that involves releasing a new version of an application to a small group of users before rolling it out to the entire user base
- New application version deployed to a temporary ASG with same capacity
- small % of traffic is sent to temporary ASG for a configurable amount of time
- deployment health is monitored
- automated rollback, if failures faced in new deployment
- Once new deployment confirmed working successfully, the instances are migrated from temporary to original ASG
- Old version is then terminated
- no downtime

#### Beanstalk Lifecycle Policy:
- Beanstalk can store 1000 application version
- If we don't remove old app versions, then it will reach 1000 versions and unable to upload new version of the app
- To phase out old app versions, we use lifecycle policy
    - based on time
    - based on space
- Versions that are used currently will not be deleted
- Option not to delete the source bindle of application in S3 to prevent data loss

#### Setup Lifecycle Policy in Elastic Beanstalk:
- Go to Elastic Beanstalk console
- Click on the Application option (left-side)
- Choose the Application for which lifecycle policy to be created
- Go to Settings option (right-side above)
    - Lifecycle
        - Lifecycle Rule
            - by total count
            - by age
    - Retention (source bundle in S3)
        - Retain source in S3 bundle
        - Delete source from S3 bundle
    - Existing Service Roles

#### Beanstalk Extensions:
- a zip file contains code must be deployed to Elastic Beanstalk
- All parameter set in the UI can be configured with code using files
- Requirement:
    - have .ebextensions/ directory in the root of source code
    - YAML/JSON format
    - .config extension (logging.config)
    - able to modify some default settings using: option_settings
    - ability add resources such as RDS, Elastic Cache, DynamoDB, etc
- Resource managed by .ebextensions get deleted if environment goes away

#### Configure Beanstalk Extensions (JSON application):
- .ebextensions/ in root folder source code
    - environment-variable.config (inside the .ebextensions folder)
- app.js
- cron.yaml
- icon.txt
- index.html
- package.json

Note:
- https://docs.aws.amazon.com/elasticbeanstalk/latest/dg/command-options-general.html
- https://docs.aws.amazon.com/elasticbeanstalk/latest/dg/ebextensions-optionsettings.html


environment-variable.config (template):

```python
""" 
option_settings:
    namespace: namespace
    option_name: option name
    value: option value
"""
```

environment-variable.config example:

```python
""" 
option_settings:
    namespace:  aws:elasticbeanstalk:container:tomcat:jvmoptions
    option_name:  Xmx
    value:  256m
    option_name: MYPARAMETER
    value: parametervalue
"""
```

environment-variable.config (shorthand template):

```python
""" 
option_settings:
  aws:elasticbeanstalk:container:tomcat:jvmoptions:
    Xmx: 256m
  aws:elasticbeanstalk:application:environment:
    MYPARAMETER: parametervalue
"""
```

environment-variable.config (shorthand):

Note:
- the changes made in .ebextensions/ folder (environment-variable.config file) will reflect in environment properties


```python
""" 
option_settings: 
    aws:elasticbeanstalk:application:environment:
        DB_URL: "jdbc:postgresql://rds-url-here.com/db"
        DB_USER: username
"""
```

environment-variable.config (JSON):

```python
{
  "option_settings": [
    {
      "namespace": "aws:elasticbeanstalk:container:tomcat:jvmoptions",
      "option_name": "Xmx",
      "value": "256m"
    },
    {
      "option_name": "MYPARAMETER",
      "value": "parametervalue"
    }
  ]
}
```

#### Elastic Beanstalk Migration:

Load Balancer:
- Once the Elastic Beanstalk environment is created, then the load balancer type cannot be modified (only the configuration of Load Balancer type)
- So, we cannot perform clone operation of Elastic Beanstalk environment

To mitigate this:
- Create a new environment with same configuration and deploy
- perform CNAME swap or Route53 update
