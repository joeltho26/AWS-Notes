#### Benefits of Cloud Formation:

- infrastructure as code:
    - full control and visibility of created AWS resource in code
    - it is version controlled using GIT

- cost:
    - cost of AWS resources is stack tagged with an identifier
    - cloudformation template could help in estimation of AWS resources cost
    - savings strategy: could do automation delete of templates at 5 PM and recreate at 8 AM

- production:
    - ability to destroy and re-create AWS resources infrastructure
    - automated generation of diagram of the cloud formation template
    - declarative programming

- separation of concern:
    - create many stacks for many apps and many layers, etc
        - VPC stacks
        - Network stacks
        - App stacks

- Don't re-invent the wheel:
    - Leverage existing CloudFormation template

- Template uploaded to S3 bucket

- CloudFormation creates the Stacks (AWS resources)
- Stacks are identifiable by name

#### CloudFormation - Rollbacks:
- Stack Creation Fails:
    - Default: everything rollbacks (gets deleted)
    - option to disable rollback and troubleshoot what happened using logs and error messages

- Stack Update Fails:
    - stack automatically rollbacks to previous version (known working state). Check logs and error messages to troubleshoot

- Rollback Fails:
    - Fix resources manually and issue ContinueUpdateRollback API from console or CLI


#### Setup Rollback CloudFormation (stack creation fails - preserve successful resource provisioned):
- Go to CloudFormation console
    - while creating a Stack, 
        - the uploaded CloudFormation template has an invalid resource details (resulting in error and rollback)
        - Go to Stack Failure option
            - Rollback all stack resources
            - preserve successful provisioned resources * (for the errored ones if there is a previous version then its rolled back to the previous version if not then the resource from the stack is deleted)
    
Note:
- We cannot update the Stack for the above use case, So we delete and recreate the Stack from start

#### Setup Rollback CloudFormation (stack update fails - rollback all resource provisioned):
- Go to CloudFormation console
    - while updating a Stack, 
        - the uploaded CloudFormation template has an invalid resource details (resulting in error and rollback)
        - Go to Stack Failure option
            - Rollback all stack resources *
            - preserve successful provisioned resources

Note:
- the resources will be rolled back to the working version

#### Setup Rollback CloudFormation (stack update fails - preserve successful resource provisions):
- Go to CloudFormation console
    - while updating a Stack, 
        - the uploaded CloudFormation template has an invalid resource details (resulting in error and rollback)
        - Go to Stack Failure option
            - Rollback all stack resources 
            - preserve successful provisioned resources *

Note:
- the resources will be rolled back to the working version, but the successfully provisioned resources will still be available
