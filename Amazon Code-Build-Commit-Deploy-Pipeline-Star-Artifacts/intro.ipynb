## CI/CD on AWS:
- automatic deployment
- By just pushing the code to github code repository, CI/CD will handle
    - managing specific AWS services 
    - Performs test to validate the change if they are working fine
    - Performs the changes in different stages for such as dev/prod/test/staging based on the code changes and request branch
- AWS CodeCommit: store the code
- AWS CodePipeline: automate our pipeline from code to Elastic Beanstalk (deploying AWS resources)
- AWS CodeBuild: build and test the code
- AWS CodeDeploy: deploy the code to EC2 instances
- AWS CodeStar: manages software development activities in one place
- AWS CodeGuru: automated code reviews using Machine Learning
- AWS CodeArtifacts: store, publish and share software development packages

#### Continuous Integration (CI):
- Developer push the code to code repository often (eg: Github, CodeCommit, BitBucket, etc)
- A testing / build server checks the code as soon as its pushed
- The developer gets feedback about the tests performed on the code pushed. Provide the code checks and fails
- Find bugs early and fix them immediately
- Deliver code changes faster since the code is tested in testing/build server and not the developers machine

#### CodeCommit:
- It is code repository to store code and track changes and rollbacks
- Version control: (GIT)
    - to understand the code changes happened over a period of time such as 
        - what were the code changes? 
        - who committed the code changes?
        - when was the code changes committed?
- Ability to create different branches to separate the various versions of code for development, test, production
- Collaborate with other developers
- Fully managed and highly available
- Increased security and compliance (encrypted, access control, etc)
- No size limit on repository
- Private Git repository

#### CodeCommit Security:
- Interactions = GIT
- Authentication = SSH Keys (generate in IAM console for users), HTTPS (AWS CLI Credentials Helper or Git Credentials for IAM user)
- Authorization = IAM policies to manage users/roles permissions to repository
- Encryption = At Rest (repositories automatically encrypted using KMS), In transit (HTTPS or SSH)
- Cross Account Access = IAM roles + AWS STS (AssumeRole API)

#### CodePipeline:
- Visual workflow to orchestrate CI/CD 
    - Source: CodeCommit, Github, GitLab, etc
    - Build: CodeBuild, Jenkins CD, CloudBees, TeamCity etc
    - Test: CodeBuild, AWS Device Farm, 3rd Party tools, etc
    - Deploy: CodeDeploy, Elastic Beanstalk, CloudFormation, ECS, S3
    - Invoke: Lambda, Step Functions
- Consist of stages:
    - Each stage has sequential and/or parallel actions
    - Example: Build -> Test -> Deploy -> Load Testing
