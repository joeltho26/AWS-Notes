#### AMI (Amazon Machine Image):
- Region specific (but can be copied across region)
- customization to EC2 instance
    - add specific software, tools, OS, configuration, monitoring, etc

#### Types of AMI:
- Public AWS AMI ((created by other Amazon))
- Custom AMI (created by an user)
- Marketplace AMI (created by other AWS user)

#### AMI Process:
- Start EC2 instance and customize it (in one AZ - us-east-1a)
- Stop the instance (data integrity)
- Build an AMI, will create EBS snapshots
- Launch instance from other AMIs (in another AZ - us-east-1b)

#### To create AMIs:
- Launch the EC2 instance
- Create an image of the instance
    - select the EC2 instance and right-click choose create image from image and templates  
        - name
        - create

#### To check AMIs created:
- In EC2 page on the right side, we have option AMI
    - We can find the AMI created in the AMI page
        - AMI id
        - image type
        - AMI name
        - platform details
        - owner, etc

#### Two ways to launch EC2 instance with AMI created:
1. By standard way of launching EC2 instance
    - Application and OS Image
        - Go to My AMI option 
            - Owned by me
                - choose the AMI
2. Go to AMI page from EC2 and launch the EC2 instance with the AMI image configuration

#### To delete AMIs:
- Terminating the EC2 instance will not delete the AMI created
- First we have to deregister the AMI
- Go to snapshot page in EC2 to delete the AMI created via AMI option

#### Recover AMI after deletion:
- By using the recycle bin option similar to EBS snapshot recycle bin
