#### Apache Spark:
- Distributed Processing Framework for big data
- In-memory caching, optimized query execution
- Supports, Java, Python, Scala, R
- Supports code re-use
    - Batch Processing
        - Batch processing is a method for computers to process large amounts of data at once. It's often used for repetitive tasks that would be inefficient to run on individual transactions
    - Interactive Queries
        - Spark SQL
    - Real-time Analytics
    - Machine Learning
        - MLLib
    - Graph Processing
        - analyzing data using graphs to identify relationships and patterns
- Spark Streaming
    - integrated with Kinesis, Kafka on EMR
- Spark is not meant for OLTP
- spark application can be hosted on EMR cluster 

Note:
- stream processing is the process of analyzing data as it flows through a system in real time

#### Spark Architecture:
- Spark apps are run as independent processes on a cluster
- The SparkContext (driver program) co-ordinates the processes
- The SparkContext works via Cluster Manager
    - incase of EMR or Hadoop, its YARN which is built-in Hadoop
    - incase of SPARK, it has built-in manager
- Executors run computation and store data
    - In EMR: 
        - a "core node" is a node that both runs tasks and stores data on the cluster
        - a "task node" only runs tasks and does not store data
- The SparkContext sends application code and tasks to executors

#### Spark Components:
- Spark CORE
    - memory management
    - fault recovery
    - scheduling
    - distribute
    - monitor jobs
    - interact with storage (HDFS, EMR FS, etc)
    - supports Python, Java, Scala, R programming languages
- Spark Streaming
    - real-time streaming analytics
    - structured streaming
    - supports sources like Kafka, Flume, HDFS, ZeroMQ, Twitter

- MLLib
    - used to apply various Machine Learning Algorithm
- Spark SQL
    - Query Engine
    - 100x faster than MapReduce
    - supports HiveQL and other formats as well Parquet, HDFS, ORC, JSON, ODBC, JDBC
    - introduces the concept of dataset or dataframe (python)
- GraphXs
    - Graph processing ETL, analysis, iterative graph computation

Note:
- Spark Core under the hood uses RDD
    - RDD stands for Resilient Distributed Dataset, which is a data structure in Apache Spark that stores data across multiple machines in a cluster. RDDs are used for parallel processing and fault tolerance

#### Spark Integration with Athena:
- can run jupyter notebooks in Athena
    - notebooks can be encrypted with KMS
    - notebooks can be later used by applications to develop visualization and analyze the data
    - notebooks can be also integrated with ML algorithms
- totally serverless
- Uses Firecracker to spin up spark resources
- Allow CLI/programmatic API 
    - create workgroups, notebooks, start-session, start-calculation-execution, etc
- can adjust the DPUs for Executor sizes and coordinators
    - set upper limits to not pay more
- Pricing based on DPUs and GPUs

Note:
- Firecracker is a virtual machine monitor (VMM) that uses the Linux Kernel-based Virtual Machine (KVM) to create and manage microVMs
