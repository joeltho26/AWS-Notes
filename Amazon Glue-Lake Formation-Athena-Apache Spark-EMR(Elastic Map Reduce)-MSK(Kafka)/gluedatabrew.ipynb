#### AWS Glue DataBrew:
- visual data preparation tool  
- UI for pre-processing large dataset
- Input Source: S3, data warehouse, database
- Output to S3
- there are more than 250 ready-made transformation that we can apply to data
- create "recipes" of transformation that can be saved as job within a larger project
    - apply multiple transformation using recipe of those transformations
- recipes made of recipe actions
- Eg: as shown below  
    - NEST_TO_MAP
        - take selected columns and convert them to key-value pairs 
        - so instead of individual columns, get converted into a JSON document which has key-value mappings
- We can define data quality rules part of DataBrew 
- create datasets with custom SQL from Redshift & Snowflake
- Security:
    - use KMS keys (at rest)
    - SSL in transit
    - IAM 
- Logging:
    - CloudWatch & CloudTrail

#### Create Recipe in DataBrew:
- Now to create recipes, go to right-side pane
    - Add Step (build your recipe)

            - s3 location

#### Handling Personally Identifiable Information (PII) in DataBrew transformation:
- Substitute with Random numbers
    - REPLACE_WITH_RANDOM
- Shuffle
    - SHUFFLE_ROWS
- Deterministic Encryption (a given same value has a same encrypted value)
    - encryption is two-way process 
- Probability Encryption (more than one result in an encryption of a given field)
- Decryption
- Null out or Delete values
- Mask out (eg: only credit card last 4 digits are displayed and none else)
- Hashing (on-way process)
