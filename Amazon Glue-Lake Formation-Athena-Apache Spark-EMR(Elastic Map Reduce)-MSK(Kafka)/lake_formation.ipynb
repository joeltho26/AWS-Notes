#### Lake Formation:
- easier to setup secure data lake within days
- loading data and monitoring data flows
- setting up partitions
- encryption and managing keys
- access control
- auditing
- define transformation jobs and monitor them
- built on top of AWS Glue

Note:
- While both AWS Lake Formation and AWS Glue are used for managing data lakes on AWS, the key difference is that Lake Formation focuses primarily on data governance and fine-grained access control, while Glue is primarily used for data integration and ETL tasks, including data discovery and transformation through crawlers and jobs; essentially, Lake Formation manages data permissions on top of the data catalog created by Glue, allowing for granular control over who can access what data within your data lake

#### Cross Account - Lake Formation permission:
- If an user from another account (Account B) wants to access the created Lake Formation (Account A)
    - Cross account Lake Formation permissions
        - Recipient must be setup as a data lake administrator
        - can use AWS Resource Access Manager for accounts external to your organization
        - IAM permissions for cross-account access
- Lake Formation does not support manifests in Athena or Redshift queries
- IAM permissions on the KMS encryption key are needed for encrypted data catalogs in Lake formation
- IAM permissions needed to create blueprints and workflow

#### Governed Tables & Security - Lake Formation:

Governed tables:
- supports "governed tables" for ACID transactions across multiple tables
    - new type of S3 table
    - cant change the choice afterwards
    - works with streaming data too (Kinesis)
    - can query with Athena
- feature of storage optimization with automatic compaction
    - the above point overcomes the drawback of ACID support on data lake
        - overhead to compress and compact the table periodically to avoid the data increasing over time

Security:
- Granular Access Control with Row & Column level security
    - for both governed tables and regular S3 tables

Note: 
- ACID transaction means strict transformation rules i.e. when we have multiple users concurrently performing operations such as reading, updating, deleting rows, ACID transactions will take care of it without worrying about the stomping of multiple users to perform concurrent actions

#### Data Permissions - Lake Formation:
