#### Microsoft Active Directory:
- Microsoft Active Directory (AD): It is a software found on any windows server with AD domain services
- Database of objects: Users, Accounts, Computers, File Shares, Security Groups
- Centralized Management system: We can create accounts, manage permissions, etc
- Objects are organized as trees
- collection of trees are called forest

#### Types of AWS Directory Services:

AWS Managed Microsoft AD:
- AD Services on AWS
- Support MFA
- Establish trust connection between on-premises AD and AWS Managed AD 
- users are managed at both on-premises AD and AWS Managed AD

AD Connector:
- Directory Gateway proxies the request to on-premises
- Supports MFS
- All the users are managed at the on-premises AD


Simple AD:
- Cannot be joined with on-premises AD
- AD compatible managed directory on AWS
