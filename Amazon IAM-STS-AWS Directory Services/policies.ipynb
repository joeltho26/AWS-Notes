## IAM Policies:

- IAM permission policy document(JSON):
    - Version (date)
    - ID (optional - strings)
    - Statement
        - Sid (optional - numeric) 
        - Effect (Allow/Deny)
        - Principal (which account user or role the policy to be applied ARNs)
            - Service
        - Action (like get, put, delete, etc)
        - Resource (s3, rds, sqs, etc ARNs)
        - Condition (condition when the policies to be applied)
            - ArnLike
                - AWS:SourceArn
            - StringEquals
                - AWS:SourceAccount

## Inline Policy vs AWS Managed Policy:
- Inline policy:
    - It a policy manually created by the user by using the JSON policy document
- AWS Managed Policy:
    - Its a AWS created policy and we just add the permission policy to an IAM user/role/user groups.
