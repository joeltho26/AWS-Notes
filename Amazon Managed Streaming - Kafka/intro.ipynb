#### Amazon Managed Streaming - Kafka:
- Its alternative to Kinesis
- MSK is fully managed Apache Kafka on AWS
    - allows to create, update, delete clusters 
    - MSK creates and manages Kafka broker nodes & Zookeeper nodes
    - Deploy the MSK cluster in VPC, multi-AZ
    - Automatic recovery from Kafka failures
    - Data stored in EBS volume
- Can build producers and consumer of data
- can create custom configuration for clusters
    - default message size 1 MB, but after custom configuration we can send larger messages ex: 10 MB

#### Apache Kafka at high level:
- The producers consume the data from various sources like Kinesis, IOT, RDS, etc and the producers write them into the Kafka topic in Broker node 1
- After writing into Kafka topic now they are replicated to other Broker nodes 2 & 3 which are in different AZ for multi AZ
- The consumed will poll for the topic and send it to other AWs services such as S3, Sagemaker, Kinesis, RDS, etc

#### MSK Configuration:
- Choose the number of AZ (3 recommended)
- choose VPC & subnets
- choose broker instance type
- choose the number of brokers per AZ
- 1 Zookeeper per AZ
- choose EBS volume between 1 GB to 16 TB

#### MSK Security:

Encryption:
- At flight encryption via TLS between brokers
- At flight encryption via TLS between clients and brokers
- At rest encryption for EBS volumes via KMS service

Network Security:
- Use Security groups to allow only specific traffic for Apache Kafka clients

Authentication & Authorization:
- Use IAM permissions for both authorization and authentication
- Use Mutual TLS (authentication) + Kafka ACLs (authorization)
- Use SASL/SCRAM (authentication) + Kafka ACLs (authorization)

#### MSK Monitoring:

Cloudwatch Metrics:
- basic monitoring (cluster & broker level metrics)
- enhanced monitoring (enhanced broker level)
- topic-level monitoring (enhanced topic level)

Prometheus (Open source monitoring):
- opens a port on the broker to export cluster, broker, and top level metrics
- setup the JMX Exporter (metrics) or Node Exporter (CPU & disk metrics)

Broker Log Delivery:
- delivery to CloudWatch Logs
- delivery to Amazon S3
- delivery to Kinesis Data streams

#### Amazon MSK Connect:
- Amazon Managed Kafka connect workers
- Kafka Connect is a framework on Kafka to take data from Kafka and put it to somewhere else
- Auto scaling capability for Amazon Connect workers
- Deploy Kafka connect connectors to MSK Connect as plugin
    - Amazon S3, Redshift, OpenSearch, Debezium, etc

#### MSK Serverless:
- Run MSK Apache Kafka without managing any capacity
- it automatically provisions resources, scales computes & storage
- Security: get IAM access control for all clusters
- we only define the topics and the partitions are good to go
- its expensive than the normal Amazon MSK service
