#### RDS (Relational Database Service):
- SQL language used for RDS
- AWS managed SQL service
- Aurora (AWS owned), MySQL, Oracle, MariaDB, Postgres, MS SQL Server, IBM DB2

#### RDS Benefits:
- Monitor Dashboard
- Continuous backups
- Automated provisioning and updates
- Read replicas
- Scaling
- Maintenance windows
- Multi AZ setup (disaster recovery)
- storage backed by EBS

Note:
- Cannot SSH into RDS instances

#### RDS Storage Auto scaling:
- helps you increase the storage of RDS instance dynamically
- set MAXIMUM STORAGE THRESHOLD value to limit the auto scaling
- Rules for auto scaling
    - auto scaling works automatically if available storage is less than 10%
    - low storage last for 5 minutes straight
    - 6 hours passed since last modification

#### RDS Read Replicas & RDS Multi AZ:

##### Read Replicas:
- setup when we have high read requests
- upto 15 Read Replicas
- Read Replicas are read-only
- within AZ or cross AZ or cross Region
- Replication is ASYNC, reads are consistent
- Replicas can be promoted as its OWN DB
- App must update connection string to leverage read replicas
