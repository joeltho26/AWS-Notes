#### Apache Flink Capabilities:
Connectors:
- The Flink connector kinesis library lets us consume the data stream from Java
- Library of "Table API" connectors are included in Flink
    - Kafka, DynamoDB, Firehose, Kinesis, MongoDB, OpenSearch, JDBC, etc
- custom sink connectors allow flink to talk to anything you have an SDK for
    - eg: write a sink to connect to Timestream database

Operators:
- Transform one or more datastreams into a new datastream
- Similar Apache Spark Transformation
    - Map, Filter, Reduce, Join, Flatmap, Windows
- Physical partitioning
- Task chaining
