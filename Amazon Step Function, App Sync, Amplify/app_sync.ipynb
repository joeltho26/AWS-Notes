## AppSync:
- It is a managed service that uses GraphQL
- GraphQL makes it easy for application to get exactly the data they need
    - It is a query language and server-side runtime for APIs
- Includes data combining from one or more sources
- Integrates with DynamoDB, Aurora, OpenSearch and others (i.e. noSQL data stores, Relational databases, HTTP APIs, etc)
- Retrieve real-time data with WebSocket or MQTT on WebSocket
- For mobile apps: local data access or data synchronization
- It all starts with uploading one GraphQL Schema

Four security option with AppSync:

#### Hands On - AppSync:
- Go to AWS AppSync console
- Create API
    - API Options
        - GraphQL APIs 
            - manages single API by providing single endpoint to query or update data securely from multiple databases, microservices, APIs, etc
        - Merged APIs 
            - manages a group of AWS AppSync GraphQL APIs owned by different teams by composing a single GraphQL Schema and authorizing data securely from multiple backend APIs via single API endpoint
    - GraphQL API data source
        - Design from scratch *
        - Import from DynamoDB
        - Create a real-time API
    - API Details
        - Name
        - Contact details
        - Private API configuration
    - Create GraphQL type
        - Create GraphQL resources later
        - Create type backed by a DynamoDB table now *
            - Model Name
            - Fields (DynamoDB)
                - id
                - name
                - age
                - certified
            - Configure Model table
                - table name
                - primary key
                - sort key
                - additional indexes
            - Create Table
