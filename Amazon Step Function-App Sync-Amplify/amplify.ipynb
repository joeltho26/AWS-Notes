## AWS Amplify:
- used to create mobile and web applications
- consist of 4 services
    - Amplify Studio (visually build full-stack app both frontend UI and backend)
    - Amplify CLI (configure Amplify backend with CLI workflow)
    - Amplify Libraries (connect app to existing AWS services)
    - Amplify Hosting (host secure, fast, reliable web apps or websites via AWS content delivery network)
- set of tools to get started with mobile and web applications (like elastic beanstalk for mobile and web applications)
- must have features such as data storage, authentication, storage, machine learning (all powered by AWS services)
- must have front-end libraries with components ready to use for React.js, Vue.js, Next.js, Javascript, Flutter, iOS, Android, etc
- Incorporates best AWS practices, such as security, readability, scalability, etc
- Build and deploy with Amplify Studio or Amplify CLI

#### Amplify - Important features:

Authentication:
- user authentication, registration, account recovery, etc
- Leverages Amazon Cognito
- supports MFA, Social Sign-in, etc
- Pre-built UI components
- Fine-grained authorization

DataStore:
- Leverages Amazon AppSync & Amazon DynamoDB
- Works with local data and automatic synchronization with cloud
- offline and real-time capabilities
- powered by GraphQL
- Visual data modeling with Amplify Studio

Amplify Hosting:
- Build and Host Modern Web Apps
- CI/CD
- Pull Request Reviews
- Custom Domains
- Monitoring
- Password Protection
- Redirect & Custom Headers
