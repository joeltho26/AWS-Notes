## VPC (Virtual Private Cloud):
- private network to deploy resources (regional)
- If we have 2 regions, then we have 2 VPCs
- Communication between VPCs are by AWS Transit Gateway or VPC Peering

Subnets:
- Subnets are at availability zones
- public subnets can access the internet using public IP or Elastic IP
- private subnets can access resources via private IP
- to define access to internet and between subnets we use routing tables (Network ACLs)

#### Internet Gateway (IGW) & NAT Gateway:
- Internet Gateway:
    - used to connect the public subnet in the VPC to the internet
- NAT Gateway:
    - Used by private subnet to connect to the internet
    - AWS managed
    - Create a NAT gateway in public subnet and then it communicates with the Internet Gateway

### Network ACLs & Security Groups
Security Groups:
- At instance level
- only allow rule
- all rules to be evaluated before access
- stateful
- include only IP & other security groups

Network ACLs
- at subnet level
- both allow and deny rules
- rules are evaluated in order
- stateless
- include only IP

Flow: (Private subnet)
- Traffic will first go to IGW and use NAT Gateway (in public subnet) to access the request
- Then, it will go to the private subnet, uses the Network ACLs to validate the request
- Then, it goes to security group and validates for the incoming traffic, as a second line of defense.

Note:
- default Network ACLs will allow every traffic in and out

### VPC Peering:
- connect VPCs privately using AWS network within same or another region
- must not have any overlapping CIDR
- connection is not transitive
    - i.e. If VPC A & VPC B is connected via VPC peering likewise, VPC A & VPC C is connected. Then it doesn't mean VPC B & VPC C are connected. We need to establish connection for all VPCs separately.

### VPC Endpoints:
- Usually AWS services (EC2 talking to RDS or another AWS resource) they communicate via public network for public subnets
- Private subnets, endpoints allow you to connect to AWS services using a private network instead a public network
- gives enhanced security and lower latency

VPC Endpoint Gateway:
- VPC endpoint gateway only for S3 and Dynamo DB

VPC Endpoint Interface:
- VPC endpoint interface for other services
